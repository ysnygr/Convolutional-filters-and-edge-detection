```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

image = cv2.imread("/home/yasin/Documents/Desktop/images/multi_faces.jpg")
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap="gray")
```

```python
face_detector = cv2.CascadeClassifier("/home/yasin/Downloads/haarcascade_frontalface_default.xml")
faces = face_detector.detectMultiScale(gray, 4, 6) 
for (x,y,w,h) in faces:
    cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.imshow(image_copy)    
```
