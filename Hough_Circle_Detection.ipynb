```python
import numpy as np
import matplotlib.pylab as plt
import cv2

image = cv2.imread("/home/yasin/Documents/Desktop/images/round_farms.jpg")
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
gray_blur = cv2.GaussianBlur(gray,(3,3),0)
plt.imshow(gray_blur,cmap="gray")
```

```python
circles =cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, 1, 
minDist=45,
param1=70,
param2=11,
minRadius=20,
maxRadius=40)


gray_blur_copy = np.copy(gray_blur)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    cv2.circle(gray_blur_copy,(i[0],i[1]),i[2],(0,255,0),2)
    cv2.circle(gray_blur_copy,(i[0],i[1]),2,(0,0,255),3)

plt.imshow(gray_blur_copy,cmap="gray")
```
